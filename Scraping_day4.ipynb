```python
#
# Library scraping
import pandas as pd
from urllib.request import urlopen  #important for opening webpages in python
from bs4 import BeautifulSoup       # this will be used to get the HTML of the webpage and work with it

# Opening HTML

url="https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;template=results;type=batting"
text= urlopen(url)
text

#Accessing html
soup=BeautifulSoup(text,"lxml") 

# Finding Data which we want to download
table= soup.find_all("table", attrs={'class':'engineTable'})    # parameters : the tag we want to locate & attributes we want to locate

#Need to manually find in HTML page differentiating tag for the data we want to download. In our case it is <caption> tag
for temp_table in table:
  caption_tag= temp_table.findAll('caption')
  if((len(caption_tag)>0)):
    int_table=temp_table           # gets us the block where our data resides.in this case caption helped capture the table to download
#Finding Rows we want to scrape from
tr_list= int_table.findAll('tr',attrs={'class':"data1"}) # data1 was checked as identifier in html code
len (tr_list)     # checking whether no of rows is fine

# Assimilating content of single row by removing tags

rowdata=[]
for td in td_list:
  td_str= str(td)
  cleantext= BeautifulSoup(td_str,"lxml").get_text()
  rowdata.append(cleantext)

# Assimilating data across rows
for tr in tr_list:
  td_list= tr.findAll('td')
  rowdata=[]
  for td in td_list:
    td_str= str(td)
    cleantext= BeautifulSoup(td_str,"lxml").get_text()
    rowdata.append(cleantext)
  print(rowdata)


```

```python
all_players=[]
for tr in tr_list:
  td_list= tr.findAll('td')
  rowdata=[]
  for td in td_list:
    td_str= str(td)
    cleantext= BeautifulSoup(td_str,"lxml").get_text()
    rowdata.append(cleantext)
  all_players.append(rowdata)
```

```python
all_players=[]
for k in range(1,5):
  url="https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;page=" + str(k)+ ";template=results;type=batting"
  text= urlopen(url)

#Accessing html
  soup=BeautifulSoup(text,"lxml") 

# Finding Data which we want to download
  table= soup.find_all("table", attrs={'class':'engineTable'})
  for temp_table in table:
    caption_tag= temp_table.findAll('caption')
    if((len(caption_tag)>0)):
      int_table=temp_table

  tr_list= int_table.findAll('tr',attrs={'class':"data1"})

  for tr in tr_list:
    td_list= tr.findAll('td')
    rowdata=[]
    for td in td_list:
      td_str= str(td)
      cleantext= BeautifulSoup(td_str,"lxml").get_text()
      rowdata.append(cleantext)
    all_players.append(rowdata)

all_players_df=pd.DataFrame(all_players)   #Dataframe creation
all_players_df.to_csv('playerperf.csv')   #get it in csv format
```
